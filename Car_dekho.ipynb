```python
cols = ['Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cols:
  df[i] = le.fit_transform(df[i])
```

```python
X = df.drop(['Selling_Price', 'Car_Name'],axis = 1)
y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
def evaluate(model,X_train,y_train,X_test,y_test):
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)


  print(f"Accuracy for your {model} is: ")
  print(f"R2 score for training {model} is {r2_score(y_train,train_pred)}")
  print(f"R2 score for testing {model} is {r2_score(y_test,test_pred)}")
  #MAE
  print(f"training MAE for {model} is {mean_absolute_error(y_train,train_pred)}")
  print(f"testing MAE for {model} is {mean_absolute_error(y_test,test_pred)}")
  #RMSE
  print(f"training RMSE for {model} is {root_mean_squared_error(y_train,train_pred)}")
  print(f"testing RMSE for {model} is {root_mean_squared_error(y_test,test_pred)}")
```

```python
from sklearn.linear_model import LinearRegression
li = LinearRegression()
```

```python
li.fit(X_train,y_train)
```

```python
evaluate(li,X_train,y_train,X_test,y_test)
```

```python

from sklearn.model_selection import GridSearchCV
```

```python
#Ridge -> Incase of overfitting
from sklearn.linear_model import Ridge
ridge = Ridge()
```

```python
ridge_param = {
    'alpha':[0.1,1,10,100,200]
}
```

```python
ridge_search = GridSearchCV(ridge,ridge_param,cv = 5)
```

```python
ridge_search.fit(X_train,y_train)
```

```python
best_ridge = ridge_search.best_estimator_
```

```python
evaluate(best_ridge,X_train,y_train,X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn = KNeighborsRegressor()
```

```python
knn_params = {
    'n_neighbors':[1,2,3,4,9,11]
}
```

```python
knn_search = GridSearchCV(kn,knn_params,cv =5)
```

```python
knn_search.fit(X_train,y_train)
```

```python
best_knn = knn_search.best_estimator_
```

```python
evaluate(best_knn,X_train,y_train,X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
dt_param = {
    'max_depth':[3,5,7,10],
    'min_samples_split':[3,10,17]
}
```

```python
tree_search = GridSearchCV(dt,dt_param,cv = 5)
```

```python
tree_search.fit(X_train,y_train)
```

```python
best_tree = tree_search.best_estimator_
```

```python
pd.DataFrame({
    'feature':X.columns,
    'importance':best_tree.feature_importances_
}).sort_values('importance',ascending = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf_param = {
    'n_estimators':[200,500,700],
    'max_depth':[3,5,7,10,15],
    'max_features':['sqrt','log2']
}
```

```python
rf_search = GridSearchCV(rf,rf_param,cv = 5)
```

```python
rf_search.fit(X_train,y_train)
```

```python
best_rf = rf_search.best_estimator_
```

```python
evaluate(best_rf,X_train,y_train,X_test,y_test)
```
